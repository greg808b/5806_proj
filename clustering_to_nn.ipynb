```python
import numpy as np
import pandas as pd
```

```python
total_min = pd.read_excel("./project_data/total_min.xlsx")
```

```python
team_list = total_min["TEAM"].unique().tolist()
```

```python
team = "_2000_LAL"
team_df = total_min[total_min["TEAM"] == team]
column_names = list(team_df.columns)
min_total = sum(team_df["MIN"].tolist())
arr = team_df.to_numpy()
for i in range(team_df.shape[0]):
    arr[i,-1] = arr[i,-1]/min_total
team_df = pd.DataFrame(arr,columns=column_names)
team_df
```

```python
data_playoffs = pd.read_excel("./teams_cluster_min_playoffs.xlsx",index_col = "TEAM")
data_playoffs = data_playoffs.drop(columns=[f"cluster{i}" for i in range(10)])
data_playoffs
```

```python
data_playoffs_fixed = {
    'TEAM': [team for team in data_playoffs.index],
    'PLAYOFF': data_playoffs['PLAYOFF'].values
}
data_playoffs = pd.DataFrame(data_playoffs_fixed)
data_playoffs
```

```python
output = pd.DataFrame(columns=["cluster0", "cluster1", "cluster2", "cluster3", "cluster4", "cluster5", "cluster6", "TEAM", "PLAYOFF"])
for team in team_list:
    team_df = total_min[total_min["TEAM"] == team]
    column_names = list(team_df.columns)
    min_total = sum(team_df["MIN"].tolist())
    arr = team_df.to_numpy()
    for i in range(team_df.shape[0]):
        arr[i,-1] = arr[i,-1]/min_total
    team_df = pd.DataFrame(arr,columns=column_names)
    row = []
    for i in range(7):
        row.append(sum(team_df[team_df["PLAYER"].isin(cluster7[i])]["MIN"].tolist()))
    row.append(team)
    row.append(data_playoffs[data_playoffs['TEAM'] == team]["PLAYOFF"].values[0])
    output.loc[len(output)] = row
```

```python
# sort output alphabetically by team column
output = output.sort_values(by="TEAM")
output
```

```python
output.to_excel("./clustering_output/nn_inputs/teams_cluster_min_playoffs_7.xlsx", index=False)
```
